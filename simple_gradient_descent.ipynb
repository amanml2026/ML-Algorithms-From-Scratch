```python
class gradient_descent_linearR:
    def __init__(self):
        self.W = None
        self.B = None

    def fit(self, x_train, y_train, w, b, iter, alpha):
        m = x_train.shape[0]

        for j in range(iter):

            dj_dw = 0
            dj_db = 0
            error = 0

            for i in range(m):
                y_pred = w * x_train[i] + b
                err = y_pred - y_train[i]

                dj_dw += err * x_train[i]
                dj_db += err
                error += err**2

            cost = (1 / (2 * m)) * error

            w = w - alpha * (dj_dw / m)
            b = b - alpha * (dj_db / m)

            print(f"ITERATION :- {j} --- w : {w}, b : {b}, cost : {cost}")

        self.W = w
        self.B = b

    def predict(self, x_test):
        return self.W * x_test + self.B

    def get_parameters(self):
        print('w :', self.W, '\n', 'b :', self.B)
```

```python
import numpy as np
x_train = np.array([10,20,30,50])
y_train = np.array([1050,2050,3050,5050])

x_test = np.array([15,20])
y_test = np.array([1550,2050])
```
