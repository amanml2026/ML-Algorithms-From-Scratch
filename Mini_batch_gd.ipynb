```python
import pandas as pd
import numpy as np
```

```python
class MBregressor:
    def __init__(self):
        self.w = None
        self.b = None

    def fit(self,x_train,y_train,alpha,epochs,batche_size):
        batches = int(x_train.shape[0]/batche_size)

        x_train = np.array(x_train)
        y_train = np.array(y_train)
        
        self.b = 0
        self.w = np.zeros(x_train.shape[1])

        for i in range(epochs):
            for j in range(batches):
                idx = np.random.randint(0, x_train.shape[0], batche_size)

                y_hat = np.dot(x_train[idx],self.w)+ self.b

                dj_db = -2*np.mean(y_train[idx]-y_hat)
                dj_dw = -2*np.dot((y_train[idx]-y_hat),x_train[idx])/batche_size

                self.b = self.b - (alpha*dj_db)
                self.w = self.w - (alpha*dj_dw)

            y_pred = np.dot(x_train, self.w) + self.b
            cost = (1/2) * np.mean((y_train - y_pred) ** 2)

            print(f"w = {self.w} , b = {self.b} and cost = {cost}")
```
