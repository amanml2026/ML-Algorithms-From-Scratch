```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
x_train = np.array([10,20,30,50,15,20])
y_train = np.array([1050,2050,3050,5050,1550,2050])
```

```python
plt.scatter(x_train,y_train,marker='x')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
def compute_cost(x,y,w,b):
    m=x_train.shape[0]
    cost=0
    for i in range(0,m):
        error = ((w*x[i]+b)-y[i])**2
        cost+=error

    total_cost = (1/(2*m))*cost

    return total_cost

```

```python
w=100
b=50

compute_cost(x_train,y_train,w,b)
```

```python
y_pred = (x_train*w)+b
plt.scatter(x_train,y_train,marker='x')
plt.plot(x_train,y_pred,color='r')
```

```python
def compute_cost_bfixed(x,y,w):
    m=x.shape[0]
    cost=0
    for i in range(0,m):
        error = ((w*x[i]+50)-y[i])**2
        cost+=error

    total_cost = (1/(2*m))*cost

    return total_cost
```

```python
import numpy as np
import plotly.graph_objects as go


w_2d = np.linspace(-200, 400, 100)

cost = []
for w in w_2d:
    c=compute_cost_bfixed(x_train,y_train,w)
    cost.append(c)

cost = np.array(cost)

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=w_2d,
    y=cost,
    mode='lines',
    name='parabola'
))

fig.update_layout(
    title='Cost Function vs W',
    xaxis_title='W',
    yaxis_title='Cost'
)

fig.show()
```

```python



w = np.linspace(-300, 500, 100)
b = np.linspace(-100, 100, 100)


W, B = np.meshgrid(w, b)


J = np.zeros_like(W, dtype=float)

for i in range(W.shape[0]):

    for j in range(W.shape[1]):
        J[i, j] = compute_cost(x_train, y_train, W[i, j], B[i, j])


# Plot
fig = go.Figure(data=[go.Surface(
    x=W,
    y=B,
    z=J
)])

fig.update_layout(
    title="Cost Function Bowl",
    scene=dict(
        xaxis_title='w',
        yaxis_title='b',
        zaxis_title='cost'
    )
)

fig.show()
```
