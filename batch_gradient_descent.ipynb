```python
import numpy as np
import pandas as pd
```

```python
"""
cost = 1/2m(y.T*Y-2y.T*X*B + B.T*x.T*X*B)

"""
class batch_GradientDescent_matrix:
    def __init__(self):
        self.B = None

    def fit(self,x_train,y_train,alpha,epochs):

        new_col_train = np.full((x_train.shape[0], 1), fill_value=1)
        x_train = np.hstack((new_col_train, x_train))
        y_train = np.array(y_train).reshape(-1,1)


        m = x_train.shape[0]
        self.B = np.random.randn(x_train.shape[1],1)
        for i in range(epochs):
            gradient = ((2/m)*x_train.T).dot(x_train.dot(self.B)-y_train)
            self.B = self.B - alpha*gradient

            cost = 1/(2*m)*((y_train.T@y_train)-(2*(y_train.T)@x_train@self.B)
                             + self.B.T@x_train.T@x_train@self.B)
            print(f'EPOCHS : {i} and cost = {cost}')

    def predict(self,x_test):

        new_col_test = np.full((x_test.shape[0], 1), fill_value=1)
        x_test = np.hstack((new_col_test, x_test))

        return x_test@self.B
    
    def get_intercept(self):
        return self.B[0]
    
    def get_weights(self):
        return self.B[1:].reshape(1,-1)[0]
```

```python
df = pd.read_csv("C://Users//Acer//OneDrive//Desktop//Datasets//Real_estate.csv",
                 usecols=['X2 house age','X3 distance to the nearest MRT station','Y house price of unit area'])
df.head()
df.columns = ['house_age','mrt_dist','price']
x = df[['house_age','mrt_dist']]
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)

from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x_train = scale.fit_transform(x_train)
x_test = scale.transform(x_test)
```

```python
gd = batch_GradientDescent_matrix()
gd.fit(x_train,y_train,alpha=0.1,epochs=30)
```
