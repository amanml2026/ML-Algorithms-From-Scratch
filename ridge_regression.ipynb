```python
import numpy as np
class RidgeRegression:
    def __init__(self,alpha):
        self.alpha = alpha
        self.W = None

    def fit(self,x_train,y_train):
        x_train = np.insert(x_train,0,1,axis=1)

        I = np.identity(x_train.shape[1])
        I[0][0] = 0 # We dont penalize the intercept value so first row would be zero.

        self.W = np.linalg.inv((x_train.T@x_train)+self.alpha*I)@x_train.T@y_train

    def predict(self,x_test):
        x_test = np.insert(x_test,0,1,axis=1)
        return (x_test@self.W)
    
    def get_Parameters(self):
        return self.W
    
```

```python
from sklearn.datasets import load_diabetes
df = load_diabetes()
x = df.data
y = df.target

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly = PolynomialFeatures(degree=3)

x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

lr = LinearRegression()
lr.fit(x_train_poly, y_train)

print(f"r2 score in test set = {r2_score(y_test, lr.predict(x_test_poly))}")
print(f"r2 score in training set = {r2_score(y_train, lr.predict(x_train_poly))}")
# Clearly overfitting because on train set r2 score is 0.877 and in test set it is -14.56.
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=3,solver='cholesky')
ridge.fit(x_train_poly,y_train)

print(f"w = {ridge.coef_} , b = {ridge.intercept_}\nr2 score = {r2_score(y_test,ridge.predict(x_test_poly))}")

# Clealrly overfitting is solved as r2 score in test set is now 0.315.
```

```python
rr = RidgeRegression(alpha=3)
rr.fit(x_train_poly,y_train)
print(f"w = {rr.get_Parameters()[1:]} , b = {rr.get_Parameters()[0]}\nr2 score = {r2_score(y_test,rr.predict(x_test_poly))}")
```

```python
print(f"difference in coef. in self made class and sklearn one is {np.round((rr.get_Parameters()[1:]-ridge.coef_),5)}")
print(f"difference in intercept in self made class and sklearn one is {np.round((rr.get_Parameters()[0]-ridge.intercept_),5)}")
```

No difference in parameters in self made class and the sklearn Ridge class on same alphas , so our class is SUCCESSFUL !!!
